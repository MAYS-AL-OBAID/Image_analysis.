```python
#################QGIS#######################
#PRINT feature name
data = [{'name': 'park', 'area': 50}, {'name':'lake', 'area':200}, {'name': 'school', 'area': 600}]

for names in data:
    print(names['name']) # in this loop to print only features name
    if (names['area']) > 50:        # in this part its to print only when the area larger than 50
        print(names['area'])
        


```

```python
#print all values of the grid and count 1

data =[
    [1,0,1],
    [0,1,0],
    [1,0,1]
]

count = 0

for x in data:
    for y in x:
        print(y)

        if y == 1:
            count= count+1

print('count 1s =', count)

        

```

```python
#skip area 0 and stop if area larger than 80
area =[20, 50, 0, 100, 10]
for x in area:
    if x == 0:
        continue

    if x > 80:
        break

    print(x)
        
```

```python
#sum each value in layer 1 with each value in layer 2
l1 = [10, 20, 30]
l2 = [5, 15, 25]

for i in l1:
    for j in l2:
        print(i + j)
        
```
