```python


from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
from astropy.visualization import simple_norm

hdul = fits.open('DATA_PATH.fits') # NGC7293
hdul.info()

image_data=hdul[0].data


#header = hdul[0].header

#print(header)

#plt.imshow(image_data)
#plt.colorbar()

#norm = simple_norm(image_data, 'log')

norm = simple_norm(image_data, 'log')

plt.imshow(image_data, cmap= 'inferno',  norm= norm)

plt.colorbar()
plt.show()


region = image_data[200:600, 250:550]
brightness=np.sum(region)
print('total brightness: ' , brightness)

max_pixel = np.max(image_data)
position = np.unravel_index(np.argmax(image_data), image_data.shape)
print('max brightness:' , max_pixel)
print('position:' , position)


threshold = np.mean(image_data) + 2 * np.std(image_data)
nebula_mask= image_data > threshold
brightness=np.sum(image_data[nebula_mask])

print(brightness)

```
