```python

from astropy import units as u
distance = 150 * u.kilometer
time = 2 * u.minute
speed = distance / time

speed_conv= speed.to(u.meter / u.second)

print(speed_conv)
```

```python

from astropy.coordinates import SkyCoord
from astropy import units as u

star = SkyCoord(ra=83.822 * u.degree, dec=-5.391 *u.degree, frame ='icrs')

galactic = star.galactic

print(galactic.l, galactic.b)
```

```python

from astropy import units as u
from astropy.coordinates import SkyCoord

starA = SkyCoord(ra=10 *u.degree, dec=20 *u.degree)
starB = SkyCoord(ra=12 *u.degree, dec=23*u.degree)

sep = starA.separation(starB)

print(sep)

```
